```python
url3 = 'https://stackoverflow.com/questions/tagged/python'
data3=urllib.request.urlopen(url3).read()
text3=data3.decode("utf-8")
soup3=BeautifulSoup(text3, 'html.parser')
```

```python
questionT_t=[]
questionT=[]
questionT=soup3.select("div#questions.flush-left div.summary h3")
for link in questionT:
#     href=link.get('href')
    questionT_t.append(link.string)
DfQT=pd.DataFrame(questionT_t)
```

```python
questionC_t=[]
questionC=[]
questionC=soup3.select("div#questions.flush-left div.summary div.excerpt")
for link in questionC:
#     href=link.get('href')
    questionC_t.append(link.string)
DfQC=pd.DataFrame(questionC_t)
```

```python
pd.concat([DfQT,DfQC],axis=1)
```

```python

links=[]
answers=[]

for link in soup3.select("div.question-summary h3 a"):
    links.append(link.attrs["href"])

for link in links:
    url4='https://stackoverflow.com'+link
    response = requests.get(url4)
    text= response.text
    soup= BeautifulSoup(text,'html.parser')
    for answer in soup.select("div.answercell.post-layout--right div.post-text p"):
        answers.append(answer.text)
answers
```
