```python
dfleft = pd.DataFrame({'KEY':['K0','K1','K2','K3'],
                      'A':['A0','A1','A2','A3'],
                      'B':['B0','B1','B2','B3']})
dfleft
```

```python
dfright=pd.DataFrame({'KEY':['K2','K3','K4','K5'],
                     'C':['C2','C3','C4','C5'],
                     'D':['D2','D3','D4','D5']})
dfright
```

```python
dfmjr=pd.merge(dfleft,dfright,on='KEY',how='left')
dfmjr
# merge: concat과 주기능이 동일
# left DF를 기준으로 merge병합
```

```python
dflri=pd.merge(dfleft,dfright,on='KEY',how='inner')
dflri
# 겹치는 key가 존재한다면, 이에 따라 조인
```

```python
dflro=pd.merge(dfleft,dfright,on='KEY',how='outer')
dflro
# key에 따라 행번호 자동배정
```

```python
dflrii=pd.merge(dfleft,dfright,left_index=True,right_index=True)
dflrii
# 행번호 기준 인덱싱 정렬
```
