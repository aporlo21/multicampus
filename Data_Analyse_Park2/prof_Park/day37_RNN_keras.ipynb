```python
timesteps = 10 # 시점
# 자연어 처리시, 문장 : 문장을 구성하는 단어의 길이
# 단어 : 단어를 구성하는 글자의 길이
# 단어의 길이를 동일한 크기로 통일시켜줘야한다 
# ==> 그렇기에 padding을 해준다 

# 주식예측모델
# timesteps = 7 : 7일동안의 주가흐름을 기초로 오늘 어떨지 예측해보자 의미를 지님 
# timesteps = 30 : 30일 동안의 주식변동을 기초로 오늘 어떨지 예측해보자

# 여러줄에 걸쳐입력되는것 = > batch size

input_dim = 4 # 단어 4개인 벡터(corpus)
hidden_size = 8 # 은닉상태의 크기 ( Cell 의 크기 )
inputs = np.random.random((timesteps, input_dim))
print(inputs)
print(inputs. shape) # shape(10, 4) 
# 동그란원이 10개 ,, 각각의 원에 4개 들어가있어
```

```python
# 초기 은닉 상태를 0 으로 초기화
hidden_state_t = np.zeros((hidden_size, ))
hidden_state_t

# (은닉층) 출력[ht] = tanh (wx * xt + wh * (ht-1) + b) 

# 가중치 
# wx : 입력에 대한 가중치 
# wx : (은닉상태의 크기 * 입력의 차원)
wx = np.random.random((hidden_size, input_dim)) # (Dh * d) (8,4)
# wh : 은닉상태 대한 가중치
# wh : (은닉상태의 크기 * 은닉상태의 크기)
wh = np.random.random((hidden_size, hidden_size)) # (Dh * Dh) # (8,8)
# b : 바이어스
# b : 은닉 상태 크기
b = np.random.random((hidden_size,)) # (Dh * d) # (8,)
print(np.shape(wx))
print(np.shape(wh))
print(np.shape(b))

# 단어개수 4
# 배치크기 1 
# Dh 4 
```

```python
total_hidden_states = []
# 메모리 셀 동작 { 출력을 입력으로 받는 작업 반복 }
for input_t in inputs : # (10, 4) # 10번의 timestep
#     print(input_t) # 입력으로 들어간다
    output_t = np.tanh(np.dot(wx, input_t) + np.dot(wh, hidden_state_t) + b) 
# 아래에서 위로 올라가 # + # 초기은닉상태
    total_hidden_states.append(list(output_t))
    hidden_state_t = output_t
    #     print(output_t)
# print(np.shape(total_hidden_states)) # (10,8)
print(total_hidden_states)
# 제일 상단 (10)을 hidden size (8) 로 도출
```

```python
text = """
멀티캠퍼스에 있는 교육생들이 공부하고 있다 \n
교육생들은 딥러닝을 하고있다 \n
결석한 교육생도 있고 지각하는 교육생도 있다 \n
"""
```

```python
from keras_preprocessing.text import Tokenizer
```

```python
t = Tokenizer()
t.fit_on_texts([text])
# 형태소분석도구 okt 패키지 이용바람 .. 해당단계에 해당 
print(t.texts_to_sequences([text]))
print(t.word_index)
```

```python
# 입력 : 멀티캠퍼스 # 단어수 : 2
# 출력 : - 에 있는 교육생들이 

# RNN 기반 전체 text 기반 모델생성시, 
# xdata = 멀티캠퍼스에 
# ydata = 에 있는 교육생들이
# 멀티캠퍼스에, 3 => 있는 교육생들이 공부하고 

vocab_size = len(t.word_index)+1

sequences = list()

for line in text.split('\n'):
#     print(t.texts_to_sequences([line])[0])
    encoded = t.texts_to_sequences([line])[0]
    for i in range(1, len(encoded)):
        sequence = encoded[:i+1]
        sequences.append(sequence)
#         print(sequence)
# 빈자리는 0 으로 채워준다 ex) [3,4,0,0,0,0]
print(len(sequences))
print(sequences)
# max(len(s)) for s in sequence
max_len = max(len(s) for s in sequences) # 최대길이 도출 
# padding    
    # 라인단위로 
```

```python
from keras_preprocessing.sequence import pad_sequences
# 빈자리 채우기 함수 
sequences = pad_sequences(sequences, maxlen=max_len, padding = 'pre')  # default
# pad_sequences(sequences, maxlen=max_len, padding = 'post') 
# 어느 길이로 맞춰서 패딩할래? 속성 : max_len

# array([[ 0,  0,  0,  0,  3,
# 출력 4 ]
```

```python
# ohe 작업
from keras.utils import to_categorical
```

```python
y=to_categorical(y, num_classes = vocab_size)
y
## 첫번째 행 = 4에 대한 ohv
## 0 번쨰 인덱스는 모두 0 표기 => 1번 인덱스부터 1이 표기됨 ... 0은 버린다
```
