```python
## gradient.descent. 하강하는 optimizer algo.
hf=x*w + b
cost=tf.reduce_mean(tf.square(hf - ytrain))
opt=tf.train.GradientDescentOptimizer(learning_rate=0.0015)
train=opt.minimize(cost)
```

```python
sess=tf.Session() # 세션 생성 
sess.run(tf.global_variables_initializer()) # 변수 초기화 
```

```python
for step in range(30001):
    cv, wv, bv, _ = sess.run([cost, w, b, train],
                            feed_dict={x:xtrain, y:ytrain})
    if step % 30000 == 0:
        print(step, cv, wv, bv)
        
        import matplotlib.pyplot as plt
        plt.plot(xtrain, ytrain, 'ro')
        plt.plot(xtrain, sess.run(w) * xtrain + sess.run(b))
        plt.show()
```
