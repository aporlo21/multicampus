```python
pd.to_numeric(tips_sub_miss['total_bill'], errors='coerce', downcast='float')
# 큰자료형을 작은 자료형으로 바꾸겠다. 
# float 32 ( 64->32 ) : 자료형 감소
```

```python
# imshow : 이미지 데이터 = 2차원  를 색상으로 표시
from sklearn.datasets import load_digits
# 수치 이미지 데이터들을 가져온다 
digits = load_digits()
len(digits)
digits.images
# key value 의 쌍이 5개 확인
digits.images[0]
# 8 x 8 data
# 0 이라고 하는 자리를 흰색 여백, 0이 아닌자리에 검정색을 칠한다
x = digits.images[8]
# 숫자를 이미지 배열을 통해 표현함
plt.imshow(x) # default
plt.imshow(x, cmap=plt.cm.bone_r) # 명확한 확인을 위해 greyscale
plt.imshow(x, cmap=plt.cm.bone_r, interpolation='nearest') # 색상 부드럽게  
```

```python
# contour () plot => 색깔을 통해 표현
# 
# 등고선 형태로 시각화 하는 방법 ( 지도 )
# 입력변수 2개(x, y) 출력변수 1개 (z) => 3 차원 자료 표현시
# contourf => 단순히 등고선만 나타낼때
def myFunc(x,y):
    return(1-x/2+x**4+y**2)*np.exp(-x**2-y**2) # 의미없는 다항식 기입
# np.exp : 자연상수 e를 말한다 
x = np.linspace(-3,3,256)
y = np.linspace(-3,3,256)
xx, yy = np.meshgrid(x,y)
# mesh grid : 등고선을 그리고자 하는 대상 데이터를 나타내는 특별한 행렬 
zz = myFunc(xx,yy)
zz
```

```python
# np.exp(0)
# np.exp(3)
plt.contourf(xx,yy,zz, alpha = 0.5) 
```

```python
plt.contourf(xx,yy,zz, alpha = 0.5, cmap='jet') 
```

```python
plt.contour(xx,yy,zz, alpha = 0.5) 
# (x, y (위치) 높이 h (m) 에 해당하는 게 z축)
```
