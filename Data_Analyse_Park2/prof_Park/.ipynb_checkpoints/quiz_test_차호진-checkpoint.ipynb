```python
# [Scikitlearn을 활용한 머신러닝]
# 코딩 문제는 하단에 기술했습니다.
# 빅데이터 딥러닝 - 차호진

# # 1. 위 문장을 모두 수행한 후, 아래와 같은 요구사항을 구현하시오.
# # 1) tips의 'sex‘ 컬럼의 타입을 문자열로 변환한 다음, info함수를 호출하여 결과를 확인하시오.
# # 2) tips의 1,3,5,7번 index행에 저장된 tip 컬럼의 값을 모두 ‘missing'으로 변경하시오

# # 3) scikitlearn에 있는 preprocessing 모듈중 표준화 및 정규화와 관련된 함수를 모두 기술하시오
# 표준화) StandardScaler().fit_transform()
# 정규화) from sklearn.preprocessing import normalize

# ## 2. 표준화와 정규화의 차이점을 설명하시오.
# # 1. 표준화(standardization)
# 각 observation이 평균을 기준으로 어느 정도 떨어져 있는지를 나타낼때 사용된다. 
# 값의 스케일이 다른 두 개의 변수가 있을 때, 
# 이 변수들의 스케일 차이를 제거해 주는 효과가 있다. 
# 제로 평균 으로부터 각 값들의 분산을 나타낸다. 
# 각 요소의 값에서 평균을 뺀 다음 표준편차로 나누어 준다.

# # 2. 정규화(normalization)
# 정규화는 데이터의 범위를 0과 1로 변환하여 데이터 분포를 조정하는 방법이다. 
# (해당 값- 최소값) / (최대값-최소값) 을 해주면 된다.

# # 3. 데이터 분석을 위해 필요로 하는 라이브러리 및 용도를 기술하시오.
# pandas 
# Pandas는 파이썬에서 사용하는 데이터분석 라이브러리로, 
# 행과 열로 이루어진 데이터 객체를 만들어 다룰 수 있게 되며 
# 보다 안정적으로 대용량의 데이터들을 처리하는데 매우 편리한 도구.

# ## [Azure 머신러닝을 활용한 기계학습]
# # 1. 분산과 공분산의 차이점을 설명하시오.
# 공분산이라 하면 분산과는 다르게
# 하나의 변수가 아닌 두 변수 사이의 관계를 나타낸다

# # 2. gapminder = pd.read_csv('../data/gapminder.tsv', sep='\t') 
# 위 코드를 수행한 후, 
# gapmider 변수에 저장된 컬럼 중 'year' 컬럼을 기준으로 'lifeExp' 컬럼값에 대한 평균을 구하고자 한다.
# 코드를 기술하시오.

# # 3. 머신러닝 학습 방법의 종류에 대해 나열하고, 설명하시오
# 지도학습) 
# 학습데이터를 이용해 데이터를 해석할수 있는 모델만들고 이를 바탕으로 
# 새로운 데이터를 추정함 // 양질의 다량의 학습데이터 준비필요
# 비지도학습(자율학습)) 
# 마치 학생스스로 학습하는것처럼 학습데이터 속의 
# 패턴이나 숨은 중요 핵심 컨셉을 찾아서 학습하는것
# ex) 과일이름에 따른분류라기보다 특징에 따라 분류하는 접근
# 강화학습) 
# 경우의 수가 너무많아 옳고 그른것에 대해 사전에 명확히 기술하기 어려운경우 적합
# ex) 알파고, 자율주행, 전략시뮬레이션, 상황에따라 다양한 대응가능한
# 응용에 적합한 방법

# # 4. 그룹화와 관련된 함수 및 형식을 기술하시오.
# group by(기준이 될 열(series))
# 기준열을 지정하여 특정열을 그룹별로 나누고,
# 각 그룹에 통계함수 적용하고,
# 최종 통계량이 산출된것은 통합하여 표시 
```
