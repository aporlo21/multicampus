```python
# day18_
import numpy as np
import pandas as pd
from jupyterthemes import jtplot
jtplot.style(theme='gruvboxd', context='paper', spines=True, 
             ticks=True, grid=True, gridlines='dotted', figsize=(6,4.5))
```

```python
# 활성화 함수 
# 계단함수
# def step_func(x):
#     if x > 0:
#         return 1
#     else:
#         return 0

def step_func(x):
    return np.array(x>0, dtype=np.int)
```

```python
step_func(5)
# step_func(-1)
```

```python
def sigmoid(x): # x는 결과가 저장됨
    return 1/(1+np.exp(-x)) # 자연상수 e load
```

```python
sigmoid(2)
```

```python
def relu(x):
    return np.maximum(0,x) # 두 인자 중 큰 값 반환
```

```python
# relu(4)
relu(-1)
```

```python
import matplotlib.pylab as plt
```

```python
x=np.arange(-5.0, 5.0, 0.1)
y1=sigmoid(x)
y2=step_func(x)
y3=relu(x)
y2
```

```python
plt.plot(x,y1)
# 시그모이드 함수 그래프 출력
```

```python
plt.plot(x,y2)
# 계단함수
# 음수 ---> 0
# 양수 ---> 1
```

```python
plt.plot(x,y3) 
# relu함수 
```

```python
A=np.array([[1,2,3],[3,4,5]]) # 2행 3열 
B=np.array([[5,6],[3,4],[1,2]]) # 3행 2열 
np.dot(A,B)
# 연산이 가능한 형태로 변환하여 작업하라 
# T (전치행렬) 활용하기 
```

```python
X=np.array([1,2]) 
X.shape 
# (2,)
W=np.array([[1,3,5],[2,4,6]])
W.shape
# (2,3)
Y=np.dot(X,W)
Y
```

```python
# exer_again
X = np.array([1.0, 0.5]) # (2,)
W1 = np.array([[0.1, 0.3, 0.5],[0.2, 0.4, 0.6]]) # (2,3)
B1 = np.array([0.1, 0.2, 0.3])
# W1.shape
A1=X@W1+B1
Z1=sigmoid(A1)
Z1
```

```python
W2=np.array([[0.1, 0.4],[0.2, 0.5],[0.3, 0.6]]) # (3,)
B2=np.array([0.1, 0.2]) # (2,)
A2=np.dot(Z1,W2)+B2
Z2=sigmoid(A2)
Z2
```

```python
W3=np.array([[0.1,0.4],[0.2,0.5]])
B3=np.array([0.1,0.2])
A3=Z2@W3+B3
Z3=sigmoid(A3)
Z3
```

```python
# softmax func
# 지수함수 구하기 
a=np.array([0.3, 2.9, 4.0])
exp_a=np.exp(a)
exp_a
# 지수함수 의 합 구하기
sum_exp_a = np.sum(exp_a)
# softmax func
sm_func = exp_a / sum_exp_a
sm_func
np.sum(sm_func)
```

```python
plt.plot(sm_func)
```
