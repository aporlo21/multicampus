```python
def word_pp(word):
    word=word.lower()
    stop_words=['a','an','the','in','with','to','for','from','of','at','on']
    symbols=[',', '.', ':', '-', '+', '/', '*', '&', '%', '[', ']', '(', ')']
    for stop_word in stop_words :
        if word != stop_words:
            word == word
        else : 
            word = ''
    for symbol in symbols:
        word.replace(symbol, '')
    return word
```

```python
file_opened=open('data/python_wikipedia.txt')
display(file_opened.readlines())
```

```python
token_inx={}
file_opened=open('data/python_wikipedia.txt')
for line in file_opened.readlines():
    line = line.strip()
    for word in line.split():
        word=word_pp(word)
        if word not in token_inx:
            if word != '':
                token_inx[word]=1
        else:
            token_inx[word]=token_inx[word]+1
```
