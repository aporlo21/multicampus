```python
# exer09_day9
import pandas as pd
import numpy as np
import json

# javascript문법 영향으로 개발된 가벼운 데이터 표현법,
# json : 데이터 교환하는 포맷으로 그 단순함과 유연함때문에 널리사용
# 특히 웹 브라우저와 웹서버 사이 데이터 교환에 많이 사용된다 
# 가장  많이 사용되는 json 포맷은 key value pair collection 
# json 문자열을 python타입 dic list tuple 등으로 변경하는 것을 
# json decoding이라하고 이 디코딩법은 json.loads() 메서드를 사용해 
# 문자열을 python 타입으로 변경하게 됨 
# parshing : 데이터 추출  

from jupyterthemes import jtplot
jtplot.style(theme='gruvboxd', context='paper', spines=True, 
             ticks=True, grid=True, gridlines='dotted', figsize=(6,4.5))
fr=open('data/data/bitly_usagov/example.txt')
fr.readline() # 한줄 씩 읽기
records=[json.loads(line) for line in open('data/data/bitly_usagov/example.txt', encoding = 'utf8')]
records[0] # 0번쨰 인덱싱 / 딕셔너리에 담긴 데이터 끌고오기 
records[0]['tz']
# records['tz'] # indices : 단서
# 가장 사용이 많은 표준시간대 도시
# 상위 10개 도시 출력
```

```python
time_zones=[rec['tz'] for rec in records if 'tz' in rec]
time_zones
```

```python
len(time_zones)
def get_counts(seq):
    counts={}
    for s in seq:
        if s in counts: 
            counts[s]+=1 # {'America/New_York':2}
        else:
            counts[s]=1 # {'America/New_York':1}
    return counts
```

```python
counts = get_counts(time_zones) # get_counts함수로 time_zones 탐색
print(counts)
#  counts 딕셔너리 안에 s라는 변수에 저장되어진 키의 존재 여부확인
```

```python
vk=[(count,tz) for tz,count in counts.items()] # 오름차순 정렬
vk
vk.sort()
print(vk[-15:])
```

```python
from collections import Counter
# 사전(dict)클래스의 하위 클래스로 리스트나 튜플에서 
# 각 데이터 등장 횟수를 사전형식으로 반환.
# Counter클래스의 most_common() 메서드는 등장횟수기반 내림차순정리
counts=Counter(time_zones) 
counts.most_common(10) # 내림차순정리 

```
