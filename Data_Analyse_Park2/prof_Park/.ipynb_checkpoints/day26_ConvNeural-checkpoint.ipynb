## 가중치 초기화방법 


```python
# 1. small random number

# => 평균 0 표준편차 0.01 가우시안분포
import tensorflow as tf
import random
import numpy as np
import matplotlib.pyplot as plt
from jupyterthemes import jtplot
jtplot.style(theme='gruvboxd', context='paper', spines=True, 
             ticks=True, grid=True, gridlines='dotted', figsize=(6,4.5))

# w = np.random.randn(input, output) * 0.01  # 표준정규분포 
```

```python
# 10개 계층에대한 시각화
# 층이 반복될수록 결과값 ( 평균, 표준편차 ) 이 0에 가까워진다
# why? ) 0.01(0에가까움) 을 계속곱하면서 xw 결과값도 0에 수렴해진다

np.random.seed(0)
D = np.random.randn(1000,500) 
hidden_layer_sizes = [500] * 10 # hidden layer 10개, 각 레이어는 500개가 차있어
Hs = {} # 빈 딕셔너리안에 Hs 값을 기입해준다

for i in range(len(hidden_layer_sizes)): # 10번 반복한다
    #     print(X.shape[1]) # X에 대해서 shape 
    X = D if i == 0 else Hs[i-1]
    fan_in=X.shape[1]
    fan_out=hidden_layer_sizes[i]
    w=np.random.randn(fan_in, fan_out)*0.01
    H=np.dot(X,w)
    H=np.tanh(H)
    Hs[i]=H
# Hs
# 해석 (입력계층 * weight 내적하는 과정)
# 2차원 행렬 생성
# HL node가 500개씩 있다. 
# HLS : 리스트로써 node가 500 개씩 있다 
# for 문
# HL 10개를 거쳐 갈때 마다 ws 계속 곱해 

print('입력계층 평균 %f and 표준편차 %f' % (np.mean(D), np.std(D)))
# 입력 계층 평균 : 0.002633 (0) 표준편차 : 0.9994 (1)
layer_means = [np.mean(H) for i, H in Hs.items()]
layer_stds = [np.std(H) for i, H in Hs.items()]

for i, H in Hs.items():
    print('히든 계층 %d 평균 %f and 표준편차 %f' %(i+1, layer_means[i], layer_stds[i]))
# 층이 열개정도되니깐 평균, 표준편차가 둘다 0 에 가까워져버림 
plt.figure()
plt.subplot(121) 
plt.plot(Hs.keys(), layer_means, 'ob-')
plt.xlabel('layer')
plt.title('layer_mean')

plt.subplot(122)
plt.plot(Hs.keys(), layer_means, 'ob-')
plt.xlabel('layer')
plt.title('layer_std')

plt.figure()
for i,H in Hs.items():
    plt.subplot(1, len(Hs), i+1)
    plt.hist(H.ravel(), 30, range=(-1,1))
    plt.xlabel('layer' + str(i+1))
plt.show()
# 표준 정규분포가 갈수록 0에 수렴하는 시각화를 보임    
# small network 에서는 적용가능 
# deeper network 에서는 적용불가
```

```python
# 2. 가우시안정규분포 법 

# 평균 : 0 
# 표준편차 : 1

np.random.seed(0)
D = np.random.randn(1000,500) 
hidden_layer_sizes = [500] * 10 # hidden layer 10개, 각 레이어는 500개가 차있어
Hs = {} # 빈 딕셔너리안에 Hs 값을 기입해준다

for i in range(len(hidden_layer_sizes)): # 10번 반복한다
    X = D if i == 0 else Hs[i-1]
    fan_in=X.shape[1]
    fan_out=hidden_layer_sizes[i]
    
    w=np.random.randn(fan_in, fan_out)
    
    H=np.dot(X,w)
    H=np.tanh(H)
    Hs[i]=H

print('입력계층 평균 %f and 표준편차 %f' % (np.mean(D), np.std(D)))
layer_means = [np.mean(H) for i, H in Hs.items()]
layer_stds = [np.std(H) for i, H in Hs.items()]

for i, H in Hs.items():
    print('히든 계층 %d 평균 %f and 표준편차 %f' %(i+1, layer_means[i], layer_stds[i]))

    plt.figure()
plt.subplot(121) 
plt.plot(Hs.keys(), layer_means, 'ob-')
plt.xlabel('layer')
plt.title('layer_mean')

plt.subplot(122)
plt.plot(Hs.keys(), layer_means, 'ob-')
plt.xlabel('layer')
plt.title('layer_std')

plt.figure()
for i,H in Hs.items():
    plt.subplot(1, len(Hs), i+1)
    plt.hist(H.ravel(), 30, range=(-1,1))
    plt.xlabel('layer' + str(i+1))
plt.show()

# 분포 결과값이 -1 oR 1 에 집중되어져있다
```

```python

# 3. Xavier initialization

np.random.seed(0)
D = np.random.randn(1000,500) 
hidden_layer_sizes = [500] * 10 # hidden layer 10개, 각 레이어는 500개가 차있어
Hs = {} # 빈 딕셔너리안에 Hs 값을 기입해준다

for i in range(len(hidden_layer_sizes)): # 10번 반복한다
    X = D if i == 0 else Hs[i-1]
    fan_in=X.shape[1]
    fan_out=hidden_layer_sizes[i]
    
    w=np.random.randn(fan_in, fan_out) / np.sqrt(fan_in)
    
    H=np.dot(X,w)
    H=np.tanh(H)
    Hs[i]=H

print('입력계층 평균 %f and 표준편차 %f' % (np.mean(D), np.std(D)))
layer_means = [np.mean(H) for i, H in Hs.items()]
layer_stds = [np.std(H) for i, H in Hs.items()]

for i, H in Hs.items():
    print('히든 계층 %d 평균 %f and 표준편차 %f' %(i+1, layer_means[i], layer_stds[i]))

    plt.figure()
plt.subplot(121) 
plt.plot(Hs.keys(), layer_means, 'ob-')
plt.xlabel('layer')
plt.title('layer_mean')

plt.subplot(122)
plt.plot(Hs.keys(), layer_means, 'ob-')
plt.xlabel('layer')
plt.title('layer_std')

plt.figure()
for i,H in Hs.items():
    plt.subplot(1, len(Hs), i+1)
    plt.hist(H.ravel(), 30, range=(-1,1))
    plt.xlabel('layer' + str(i+1))
plt.show()

# 분포 결과값이 보기에 정상적이다 
# 문제점 활성화함수 tanh -> relu 로 바꾸면 정상 표현이 안됨 
```

## 해당 여러 초기화방법을 모델에 적용해보자 ! 

```python
# xabier 활용하기
from tensorflow.examples.tutorials.mnist import input_data
mnist = input_data.read_data_sets('MNIST_data/', one_hot=True)
```

```python
# mnist
learning_rate=0.001
training_epochs=15
batch_size=100
# tensor 가 구성되어지는 과정을 그래프로 그린다 ( 분산처리 설계 )
# 외부 컴퓨터 실행 그래프를 이쪽에서 다시 전달받는다 
# 이 과정(정보그래프를 전달하고 전달받는 일련의 과정) 을 담당하는게 session  
# 해당 tensor를 포함하는 session 객체를 원격에서 실행해야 값을 볼수 있다 
# 그래프를 정의 : 노드(node=tensor) & 간선(edge) 

tf.reset_default_graph() # 그래프에 있는 모든 tensor 초기화해주는 함수 # 항시먼저 선언
x = tf.placeholder(tf.float32, [None,784])
y = tf.placeholder(tf.float32, [None,10])
w1 = tf.get_variable('w1', shape=[784, 256], 
                     initializer = tf.contrib.layers.xavier_initializer()) # 변수 tf.Variable 과 같다
# 만든 xavier 함수 적용된다 
# RBM 볼츠만 머신 기반으로 적절한 weight 초기화 가 이루어져
# 연산량 적으면서도 좋은모델이 적용됨, 초기화 이후 b1 
b1 = tf.Variable(tf.random_normal([256]))
L1 = tf.nn.relu(tf.matmul(x,w1)+b1)

w2 = tf.get_variable('w2', shape=[256, 256], 
                     initializer = tf.contrib.layers.xavier_initializer()) 
b2 = tf.Variable(tf.random_normal([256]))
L2 = tf.nn.relu(tf.matmul(L1,w2)+b2)

w3 = tf.get_variable('w3', shape=[256, 10], 
                     initializer = tf.contrib.layers.xavier_initializer()) 
b3 = tf.Variable(tf.random_normal([10]))

hf = tf.matmul(L2,w3)+b3

cost=tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits
               (logits=hf, labels=y))
# 바로 확률로 구한다(logits = hf
# gradient descent 대신 adam OPT 사용
train = tf.train.AdamOptimizer(learning_rate=learning_rate).minimize(cost)
```

```python
# 모델 생성

sess=tf.Session()
sess.run(tf.global_variables_initializer())
for epoch in range(training_epochs): #5만개 이미지 * 15번 트레이닝
    avg_cost=0
    total_batch=int(mnist.train.num_examples/batch_size)
    for i in range(total_batch): #500번(1번에 100개씩 이미지를 읽겠다)
        batchxs,batchys=mnist.train.next_batch(batch_size)
        myfeed={x:batchxs, y:batchys}
        cv,_=sess.run([cost, train], feed_dict=myfeed)
        avg_cost+=cv/total_batch #  100데이터코스트/500
    print('cost:','{:.9f}'.format(avg_cost))

c_pre=tf.equal(tf.argmax(hf, 1), tf.argmax(y,1))
acc=tf.reduce_mean(tf.cast(c_pre, tf.float32))
print('acc:', sess.run(acc, feed_dict=
                       {x:mnist.test.images, 
                        y:mnist.test.labels}))
```

```python
# 세션 생성
sess=tf.Session()
sess.run(tf.global_variables_initializer())
for epoch in range(training_epochs):
    avg_cost=0
    total_batch=int(mnist.train.num_examples/batch_size)
    for i in range(total_batch):
        batchxs,batchys=mnist.train.next_batch(batch_size)
        myfeed={x:batchxs, y:batchys}
        cv,_=sess.run([cost, train], feed_dict=myfeed)
        avg_cost+=cv/total_batch
    print('cost:','{:.9f}'.format(avg_cost))

c_pre=tf.equal(tf.argmax(hf, 1), tf.argmax(y,1))
acc=tf.reduce_mean(tf.cast(c_pre, tf.float32))
print('acc:', sess.run(acc, feed_dict=
                       {x:mnist.test.images, 
                        y:mnist.test.labels}))
```

```python
 
tf.reset_default_graph() #그래프에 있는 모든 텐서를 초기화
x=tf.placeholder(tf.float32, [None,784])
y=tf.placeholder(tf.float32, [None,10])

keep_prob=tf.placeholder(tf.float32)

w1=tf.get_variable("w1", shape=[784,512],initializer=tf.contrib.layers.xavier_initializer())
b1=tf.Variable(tf.random_normal([512]))
L1=tf.nn.relu(tf.matmul(x,w1)+b1)
L1=tf.nn.dropout(L1, keep_prob=keep_prob)

w2=tf.get_variable("w2", shape=[512,512],initializer=tf.contrib.layers.xavier_initializer())
b2=tf.Variable(tf.random_normal([512]))
L2=tf.nn.relu(tf.matmul(L1,w2)+b2)
L2=tf.nn.dropout(L2, keep_prob=keep_prob)


w3=tf.get_variable("w3", shape=[512,512],initializer=tf.contrib.layers.xavier_initializer())
b3=tf.Variable(tf.random_normal([512]))
L3=tf.nn.relu(tf.matmul(L2,w3)+b3)
L3=tf.nn.dropout(L3, keep_prob=keep_prob)


w4=tf.get_variable("w4", shape=[512,512],initializer=tf.contrib.layers.xavier_initializer())
b4=tf.Variable(tf.random_normal([512]))
L4=tf.nn.relu(tf.matmul(L3,w4)+b4)
L4=tf.nn.dropout(L4, keep_prob=keep_prob)


w5=tf.get_variable("w5", shape=[512,10],initializer=tf.contrib.layers.xavier_initializer())
b5=tf.Variable(tf.random_normal([10]))
hf=tf.matmul(L4,w5)+b5

cost=tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits
               (logits=hf, labels=y))
train=tf.train.AdamOptimizer(learning_rate=learning_rate).minimize(cost)
```

```python

sess=tf.Session()
sess.run(tf.global_variables_initializer())
for epoch in range(training_epochs):
    avg_cost=0
    total_batch=int(mnist.train.num_examples/batch_size)
    for i in range(total_batch):
        batchxs,batchys=mnist.train.next_batch(batch_size)
        myfeed={x:batchxs, y:batchys, keep_prob:0.7}
        cv,_=sess.run([cost, train], feed_dict=myfeed)
        avg_cost+=cv/total_batch
    print('cost:','{:.9f}'.format(avg_cost))

c_pre=tf.equal(tf.argmax(hf, 1), tf.argmax(y,1))
acc=tf.reduce_mean(tf.cast(c_pre, tf.float32))
print('acc:', sess.run(acc, feed_dict=
                       {x:mnist.test.images, 
                        y:mnist.test.labels,
                       keep_prob:1}))
```

## mnist model based CNN

```python
 # CNN 구현
    # shape 주는 부분이 까다롭다
learning_rate=0.001
training_epochs=15
batch_size=100

tf.reset_default_graph() #그래프에 있는 모든 텐서를 초기화

x=tf.placeholder(tf.float32, [None, 28*28])
ximg=tf.reshape(x, [-1, 28, 28, 1])#img 28*28*1(black/white)
y=tf.placeholder(tf.float32, [None, 10])

w1=tf.Variable(tf.random_normal([3,3,1,32]))
L1=tf.nn.conv2d(ximg, w1, strides=[1,1,1,1],padding='SAME')
L1=tf.nn.relu(L1)
L1=tf.nn.max_pool(L1, ksize=[1,2,2,1], strides=[1,2,2,1], padding='SAME')
# L1 이미지 shape (?,28,28,1)
# conv 통해 ( ?,28,28,32 )
# pooling (?,14,14,32) 

# 출력결과가 본래 이미지 ( 28*28 ) => 3,3 가 된다
# 초기화 
w2=tf.Variable(tf.random_normal([3,3,32,64]))
L2=tf.nn.conv2d(L1, w2, strides=[1,1,1,1],padding='SAME')
L2=tf.nn.relu(L2)
L2=tf.nn.max_pool(L2, ksize=[1,2,2,1], strides=[1,2,2,1], padding='SAME')
#conv2d (?,14,14,64)
#relu (?,14,14,64)
#max_pool (?,7,7,64)
L2_flat=tf.reshape(L2, [-1, 7*7*64])
#reshape => (?, 7*7*64)


w3=tf.get_variable("w3", shape=[7*7*64, 10],initializer=tf.contrib.layers.xavier_initializer())
b=tf.Variable(tf.random_normal([10]))
logits=tf.matmul(L2_flat, w3)+b
cost=tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits=logits, labels=y))
train=tf.train.AdamOptimizer(learning_rate=learning_rate).minimize(cost)

# 해석
# 28 * 28 => 3 * 3 컨볼루션 32개 => [ 2*2 풀링 14*14 32개 ]
# [ 2*2 풀링 14*14 32개 ] => 두번째 계층의 입력으로 들어간다 ( L2 (3*3))
# ( L2 (3*3)) filter 64개 => [2*2 풀링 7*7 64개]

```

```python
sess=tf.Session()
sess.run(tf.global_variables_initializer())
for epoch in range(training_epochs):
    avg_cost=0
    total_batch=int(mnist.train.num_examples/batch_size)
    
    for i in range(total_batch):
        batchxs,batchys=mnist.train.next_batch(batch_size)
        myfeed={x:batchxs, y:batchys}
        cv,_=sess.run([cost, train], feed_dict=myfeed)
        avg_cost+=cv/total_batch
    print('cost:','{:.9f}'.format(avg_cost))

c_pre=tf.equal(tf.argmax(logits, 1), tf.argmax(y,1))
acc=tf.reduce_mean(tf.cast(c_pre, tf.float32))
print('acc:', sess.run(acc, feed_dict=
                       {x:mnist.test.images, 
                        y:mnist.test.labels,
                       }))
```
