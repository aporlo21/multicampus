```python
## 상관관계 높은 순서대로 배열후 상위 10개의 피처 인덱스만 추출
# top_20_featureIdx = np.argsort(df2.corr()[1])[::-1].reset_index(drop = True)[:20]
 
df3 = df2.iloc[:, top_10_featureIdx] 
df3
df3.shape
# 해당열의 값을 기준으로 열 정렬 
```

```python
## 열 이름 초기화 0번이 타겟
df3.columns = [ i for i in range(df3.shape[1])]
df3.columns
# 열 범위( 0 - 9 ) 기준 i for 문 
```

```python
## 정규화 및 전치
from sklearn.preprocessing import minmax_scale, StandardScaler, RobustScaler
 
# xdata = minmax_scale( df3.iloc[:,1 : ] )  #결과 : 0.68xxx
# print(xdata) # 1행 2열 
# xdata = RobustScaler().fit_transform(df3.iloc[:,1 : ]) #결과 : 0.63xxx

xdata = StandardScaler().fit_transform(df3.iloc[:,1 : ])
print(xdata)
ydata = df3.iloc[:,0]
print(ydata)

 
```

```python
xdata = xdata.T
print(xdata)
ydata = np.array(ydata).reshape((1, -1))
print(ydata)
xdata.shape, ydata.shape
```

```python
## 입력 제외 레이어의 갯수
n_h = 1
 
## x의 피처의 갯수.
n_x = 9
 
## 출력층 노드의 갯수.
n_y = 1
 
## 관측치의 갯수
m = 569 # 미사용
 
```

```python
X  = tf.placeholder(tf.float64, shape = [ n_x, None ] )
y  = tf.placeholder(tf.float64, shape = [  1 , None ] )

# 입력으로 사용할 데이터의 타입만 지정해주고  
# 실제값은 나중에 세션에서 실행될때 입력해줍니다.
# 동적, 유동적 

# variable에 초기값 넘겨준다 
# 상수나 랜덤으로 초기화하기 위한 함수 random.normal
W = tf.Variable( tf.random.normal( [ 1, n_x ], dtype = tf.float64), dtype = tf.float64 )
B = tf.Variable( tf.random.normal( [ 1      ], dtype = tf.float64), dtype = tf.float64 )
## 1번째 레이어의 입력 계산식 Z
Hx = W @ X + B
 
## 1번째 레이어의 엑티베이션 펑션
A1 = tf.sigmoid( Hx )
 
```

```python
## loss(A1, y)의 총합의 평균
# cost  = -tf.reduce_mean( y * tf.log( A1 ) + (1 - y) * tf.log( 1 - A1 ) )
 
cost  = tf.losses.sigmoid_cross_entropy(y, A1)
# 딥러닝 모델의 손실함수 
train = tf.train.GradientDescentOptimizer(0.01).minimize( cost )
# cost값 최소화하는 0.01 lr을 가진 GDO를 train model로 사용
predicted = tf.cast( A1 > 0.5, dtype = tf.float64 ) 
# 텐서를 새로운 형태로 캐스팅하는데 사용, 임계치 기준 넘으면 실수 표현 
accuracy  = tf.reduce_mean(tf.cast(tf.equal(predicted ,y)  , dtype = tf.float64))
 
m = xdata.shape[1]
indices = np.random.permutation(m)
 
```

```python
## k = 7 k-fold validation ( 교차검증 )
k = 7
acc = []

for i in range(k):
    test_indices  = indices[ int( i * ( m / k )) : int((i + 1) * ( m / k ) - 1) ]
    train_indices = np.setdiff1d( indices, test_indices )

    with tf.Session() as sess:
        sess.run(tf.global_variables_initializer())
        for step in range(10000):
            cv, _ = sess.run( [ cost, train ], feed_dict = { X : xdata[:,train_indices], 
                                                           y : ydata[:,train_indices] } )
            if step % 2000 == 0:
                print(step, cv)
                    
        preds = sess.run( accuracy, feed_dict = { X : xdata[:,test_indices], 
                                                y : ydata[:,test_indices] } )  
        acc.append(preds)
        
print( 'acc : ', np.sum(acc) / k ) 

## acc :  0.9341490299823632
 
```
