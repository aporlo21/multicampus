```python
# day17_quiz01
import numpy as np
from math import sqrt
```

```python
# 문제 1
# y에서 yhat까지의 비용을 계산하는 cost 함수를 만드세요.
# def costToYhat(y, yhat):
# y(실제값)=3, 7, 10, 20
# yhat(예측값)=4, 13, 12, 30
```

```python
def costToYhat(y,yhat):
    n = len(y)
    y = np.array(y)
    yh = np.array(yhat)
    
    sum_cost=np.sum(pow((y-yh),2)) 
    cost=1/n*(sum_cost)
    cost=sqrt(cost)
    
    return cost
    
    
y=[3,7,10,20]
yhat=[4,13,12,30]
print('costFunc:', costToYhat(y,yhat))
    
```

```python
def sigmoid(val):
    e_sig = pow(np.e, -val)
    result = 1 / (1 + e_sig)
    return result

for i in [-100, -10, 0, 10, 100]:
    print('sigmoid({}):{}'.format(val,sigmoid(i)))
    
```
