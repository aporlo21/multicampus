```python
# print(ord('a'))
# print(chr(97))

from sklearn.model_selection import train_test_split

label = []
data = []
for row_index, row in mr.iterrows(): # row에 한줄의 데이터가 들어가있다 
    label.append(row.ix[0])
    row_data=[]
    for v in row.ix[1:]:
        row_data.append(ord(v)) # 문자에 대해서 아스키코드값 a - 97, b -98  # 저장시 수치를 저장
        # 한행에 대해 아스키코드값인 수치로 변환되어 리스트에 저장된채로 빠져나온다 
    data.append(row_data)
# print(data)
#         print(row_data)
    
#     print(row.ix[0]) #1번 열부터 마지막까지
# label

```

```python
data_train, data_test, label_train, label_test = train_test_split(data, label)
```

```python
(8124, 23)
len(data_train) # 6093
len(data_test) # 2031

from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
# data training start
clf=RandomForestClassifier()
clf.fit(data_train, label_train) # 모델 만드는 함수 fit 
```

```python
predict=clf.predict(data_test)
predict # 예측
# 실제 
# 실제와 예측간 정확도
ac_score=metrics.accuracy_score(label_test, predict)
ac_score
rep=metrics.classification_report(label_test, predict)
print(rep)
```
