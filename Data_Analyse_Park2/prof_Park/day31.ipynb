```python
data['Species']=data['Species'].replace(['Setosa','Versicolor','Virginica'],[0,1,2])
data['Species'].value_counts().plot(kind='bar')
```

```python
# 독립변수 / 종속변수로 분리작업 진행 
datax=data[['sepal.length','sepal.width','petal_length','petal_width']]
datay=data['Species'].values
print(data[:5])
print(data[:5])
```

```python
from sklearn.model_seleciton import train_test_split
(xtrain, xtest, ytrain, ytest) = train_test_split(detax, detay, train_size=0.8, 
                                                  random_state=True)
```

```python
# 원핫 인고딩을 해주기위해 \
from keras.utils import np_utils 
ytrain = np_utils.to_categorical(ytrain)
ytest = np_utils.to_categorical(ytest)
print(ytrain[:5])
print(ytest[:5])
```

```python
# 소프트 맥스 회귀
from keras.models import Sequential
from keras.layers import Dense
from keras import optimizers
```

```python
model=Sequential()
model.add(Dense(3, input_dim= 4, activation='softmax'))
sgd=optimizers.SGD(lr=0.01)   
model.compile(loss='categorical_crossentropy', optimizer', metrics=['accuracy'])
history = model.fit(xtrain, ytrain, batch_sieze= 1, epochs =  200, 
                    validataion_data=xtest, ytest)
```
