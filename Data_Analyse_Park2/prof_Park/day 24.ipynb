```python
import tensorflow as tf 
import pandas as pd
```

```python
def MyScaler(data): # MinMaxScaling
#     print(np.min(data)) # 전체 data에서 최소값 도출 
#     print(np.min(data, axis=0)) # 열단위 최소값
#     print(np.min(data, axis=1)) # 행단위 최소값
    de = np.max(data, axis=0)-np.min(data, axis=0)
    num = data - np.min(data, axis=0)
    return num/de # 정규화가 되어진 결과값이 도출되어진다 
    
    
    
```

```python
xy=MyScaler(xy) # 각각의 feature 에따라 분류하므로 열단위 최소값추출문을 써야한다 
print(xy)
```

```python
# xy.shape # 8,5
xdata=xy[:,0:-1] # 전체행 # 맨 처음부터 마지막 에서 두번째 
xdata.shape 
```

```python
ydata=xy[:,[-1]]
ydata
```

```python
x=tf.placeholder(tf.float32, shape=[None, 4])
y=tf.placeholder(tf.float32, shape=[None, 1])
w=tf.Variable(tf.random_normal([4,1])) # 종가확인 
b=tf.Variable(tf.random_normal([1]))
```

```python
hf=tf.matmul(x,w)+b # 가설함수 예측값 
cost=tf.reduce_mean(tf.square(hf-y))
train=tf.train.GradientDescentOptimizer(0.1).minimize(cost)
```

```python
# optimizing process
train=tf.train.GradientDescentOptimizer(1e-3).minimize(cost)
```

```python

sess=tf.Session()
sess.run(tf.global_variables_initializer())
for step in range(101):
    cv, hv, _ = sess.run([cost, hf, train], feed_dict = {x:xdata, y:ydata})
    print(step, 'cost:', cv, 'hf:', hv)
```

```python
xy=np.loadtxt('data/diabete/zoo.csv', delimiter=',',dtype=np.float32)
xy.shape #
xdata = xy[:,0:-1]
ydata = xy[:,[-1]]
```

```python
x=tf.placeholder(tf.float32, [None, 16])
y=tf.placeholder(tf.int32, [None, 1]) 
# ohe 는 type 이 int형이여야 실행된다
# 0-6 # 분류기는 7개 ( 종류 )
# w1x1 + w2x2 + ... w7x7 + b => softmax => 0번 종류의 동물에 대한 확률 => 0 번 종류 동물에 대한확률 
# w1x1 + w2x2 + ... w7x7 + b => softmax => 0번 종류의 동물에 대한 확률 => 1 번 
# w1x1 + w2x2 + ... w7x7 + b => softmax => 0번 종류의 동물에 대한 확률 
# ....
# w1x1 + w2x2 + ... w7x7 + b => softmax => 0번 종류의 동물에 대한 확률  => 6번 
# 1000000, 0100000,... => [None, 7]
```

```python
# tf ) ohe 함수 따로 존재
y_one_hot=tf.one_hot(y, 7) # 동물번호 : 3 -> 0001000 < ohe >
print('one_hot', y_one_hot)
y_one_hot=tf.reshape(y_one_hot, [-1,7])
print('one_hot after reshape', y_one_hot)
# ohe hot 함수는 한 차원 높게 변환됨 
# y : [[0],[3], .... [5]]] => [None,1] shape
# one_hot = > [[[1000000]], [[0001000],,,,[00000100]]] => [None, 1, 7]
# 우리가 바라는 출력 결과의 모습은 ( shape ) = [None,7]
# reshape을 하면됨 tf.shape(y_one_hot, [-1,7 ]) 코딩하면 
# [None, 7] 로 reshape됨
# [[[1000000]], [[001]]]
```

```python
w=tf.Variable(tf.random_normal([16, 7]))
b=tf.Variable(tf.random_normal([7]))
#x:[None,16], w:[16,7]
logit= tf.matmul(x,w)+b
hf=tf.nn.softmax(logit)
cost=tf.nn.softmax_cross_entropy_with_logits(logits=logit,
                                             labels=y_one_hot)
cost2=tf.reduce_mean(cost)
train=tf.train.GradientDescentOptimizer(0.1).minimize(cost2)
prediction=tf.argmax(hf,1)
correct_prediction= tf.equal(prediction,tf.argmax(y_one_hot,1))
accuracy=tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for step in range(10001):
        sess.run(train, feed_dict={x:xdata, y:ydata})
        if step % 100==0:
            cv, av=sess.run([cost2, accuracy], feed_dict={x:xdata, y:ydata})
            print("step:", step, "cost:",cv, "acc:", av)
#전체 데이터로 트레이닝 수행하여 모델 생성
#모델에 전체 데이터를 집어 넣어서 정확도 출력
#데이터를 분할하지 않았음
    pred=sess.run(prediction, feed_dict={x:xdata})
    for p,y in zip(pred, ydata.flatten()):
        print('[{}] 예측: {} 실제 y:{}'.format(p==int(y), p, int(y)))
            
#     print(pred)
```
