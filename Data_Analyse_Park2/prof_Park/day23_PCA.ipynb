```python
from sklearn.preprocessing import StandardScaler
x_std=StandardScaler().fit_transform(X) # 표준화
print(x_std)

```

```python
# 중심화 진행과정 
# 각 열의 평균을 0으로 만들어준다
# ==> covariance ( 공분산 )
# 공분산행렬
# feature가 col에 있는데 row로 바꾸기 위해 변환
features=x_std.T
features
# 각각의 하나 : 개인데이터 
cov_matrix=np.cov(features)  # coveriance # 5개의 features 에 대한 공분산값
print(cov_matrix)

```

```python
# 공분산 행렬로부터 고유벡터와 고유값 추출 (eigen vector, value)
eig_vals, eig_vecs = np.linalg.eig(cov_matrix)
eig_vals, eig_vecs
print('고유벡터 : %s' %eig_vecs)
print('고유값 : %s' %eig_vals)
```

```python
# 공분산행렬 내적 : 두 벡터의 유사도 표현
# 고유벡터 : 해당행렬이 벡터에 적용하는 힘의 방향
# 공분산행렬의 고유벡터는 ? 
# ans) 데이터가 어느 방향으로 
# 분산되어있는지 나타냄
# 고유값? 고유벡터에 해당하는 상관계수
# 공분산행렬의 고유값? 각 축 에 대한 공분산 값
# 결국 고유값이 큰 숫서대로 고유 벡터를 나열하면.
# 중요한 주성부터 추출이 됌

# a x rambda x
# 21 0.7 e 0.7
# eig_  sum(eig_vals)
projected_X = x_std.dot(eig_vecs.T[0])
projected_X
#  1차원 축소된것 
res=pd.DataFrame(projected_X, columns=['PC1'])
res['yaxis']=0.0
res['label']=Y
```

```python
# 표준화
X_std = StandardScaler().fit_transform(X)
```

```python
# 공분산행렬
mean_vec = np.mean(X_std, axis=0)
mean_vec # 평균

cov_mat=(X_std-mean_vec).T.dot((X_std-mean_vec))/(X_std.shape[0]-1)
cov_mat
```

```python
# print(cov_matrix)
```

```python
sns.heatmap(cov_mat, cmap = 'cubehelix', annot = True, square=True)
# 고유값, 고유벡터 추출하기
```

```python
eig_vals, eig_vecs = np.linalg.eig(cov_mat)
print('아이겐벡터 : \n %s' %eig_vecs)
print('아이겐벨류: \n %s' %eig_vals)
```

```python
# 주성분에 해당하는것을 추출할때는 위의 추출값을 내림차순 정렬으로 하여 보는것이 편하다 
eig_pairs = [ (eig_vals[i], eig_vecs[:,i] )
             for i in range(len(eig_vals)) ] # 0-6까지의 값을 갖는 i
eig_pairs
# 1인수 eigvals // 2 - eigvecs
eig_pairs.sort(key=lambda x:x[0], reverse = True) # value들에대해 역순 설정 
eig_pairs

for i in eig_pairs:
    print(i[0]) # eigval들만 추출
    
tot = sum(eig_vals)
var_exp = [(i/tot)*100 for i in sorted(eig_vals, reverse=True)]
var_exp
# eigval에 따라 내림차순정리 
# 의미 ) 데이터를 어느정도 표현할수 있느냐 
```
