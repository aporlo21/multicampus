```python
# 1. 여러 개의 파일(for문 사용)로드

with open('data/raw_data_urls.txt', 'r') as data_urls:
    for line, url in enumerate(data_urls):
        if line == 4:
            break
        fn=url.split("/")[-1].strip()
        print(url)
        fp=os.path.join('','data',fn)
        urllib.request.urlretrieve(url, fp)
            
```

```python
# 2. melt -> pivot_table 행/열 요소 변경 연습

weather_csv = pd.read_csv("data/weather.csv")
weather_csv.info()
weather_csv.iloc[:5,:10]
# weather_melt = pd.melt(ebola_csv, )
```

```python
weather_melt=pd.melt(weather_csv, id_vars = ['id','year','month','element'],
                    var_name='day')
weather_melt
```

```python
weather_pivotT = weather_melt.pivot_table(
index=['id','year','month','day'],
columns='element',
)
weather_pivotT
```

```python
# 3. Date Day variable     value status country.stataus

ebola=pd.read_csv('data/country_timeseries.csv')

ebola.columns
ebola
```

```python

new_df_ebola = pd.DataFrame({'Data':ebola.Date,'Day':ebola.Day})
ebola_sort = pd.melt(ebola, id_vars=['Date','Day'])

varify_split = ebola_sort.variable.str.split("_")

value = ebola_sort.value
variable = ebola_sort.variable
status_value = varify_split.str.get(0)
country_value = varify_split.str.get(1)

pd.concat([new_df_ebola, value, variable, status_value.rename('status'), 
           country_value.rename('country')], axis=1)

# status_value
# country_vale
# ebola_pivotT = ebola_sort.pivot_table(
# index=['Date','Day','country_value','status_value'],
# columns='variable','value'
# )
# ebola_pivotT
```
