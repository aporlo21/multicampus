```python
# 위키피디아 에서 python 키워드 검색하여 나온결과 파일 

# 전처리(소문자, stopword, 컴마, 대괄호, 특수문자 ,,, 숫자)
def word_preprocess(word):
    word=word.lower() # 소문자 처리
    # 3
    stop_words = ['a', 'an', 'the', 'in', 'with', 'to', 'for', 'from', 'of', 'at', 'on', 'until', 'by', 'and', 'but', 'is', 'are', 'was', 'were', 'it', 'that', 'this', 'my', 'his', 'her', 'our', 'as', 'not'] # make your own list
    symbols = [',', '.', ':', '-', '+', '/', '*', '&', '%', '[', ']', '(', ')'] # make your own list# 불용어 처리 사전 불러오기 or 직접 만든 구문 불러오기
    # 전치사 or 부호 list : 구문 의미에 큰 역할 없으므로 삭제
    for stop_word in stop_words : 
        if word != stop_word: # word 에 들어와 있는 단어 
            word = word
        else: 
            word = ''
    for symbol in symbols: # symbols에 들어와 있는 단어 차례로 불러와
        word.replace(symbol, '') # 겹치는 word & symbol은 날려버려
    return word
    
        
```

```python
file_opened=open('./data/python_wikipedia.txt')
display(file_opened.readlines())
```

```python
token_inx={}
file_opened=open('./data/python_wikipedia.txt')
for line in file_opened.readlines():
    # 공백제거 
    line = line.strip() # 좌우의 공백제거
    #print(line)
    for word in line.split():
        word=word_preprocess(word)
        if word not in token_inx: #word가 token_inx 딕셔너리에 없다면 이라는 의미
            if word != '':
                #token_inx[word]=len(token_inx)+1 어가 key로 작용
                # 키의개수 + 1 = 새로 들어온 단어의 개수 
                token_inx[word]=1 # 1로 초기화한다
        else: # 딕셔너리에 기존에 존재하던 key에 해당하는 값에 누적시킨다
            token_inx[word]=token_inx[word]+1

        #print(word)
# 단어 단위로 몇개 있는지 확인
# {python : 200, programming : 50,...}
# 각각의 단어별로 몇개 씩 존재하는지 값으로 딕셔너리 구성하고자함
```
