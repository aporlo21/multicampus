```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os 
import urllib.request
import glob
from jupyterthemes import jtplot 
jtplot.style(theme='gruvboxd', context='paper', spines=True, 
             ticks=True, grid=True, gridlines='dotted', figsize=(6,4.5))
```

```python
with open('./data/raw_data_urls.txt', 'r') as data_urls:
    for line, url in enumerate(data_urls):
        if line==10:
            break
        filepath = os.path.join('./data', url.split('/')[-1].strip())
        urllib.request.urlretrieve(url, filepath) 
        

df_list = []
file_list = glob.glob('./data/fhv_tripdata_2015-*')

for file in file_list:
    df_list.append(pd.read_csv(file))

taxi = pd.concat(df_list)
```

```python
ebola = pd.read_csv('./data/country_timeseries.csv')

ebola_melt = pd.melt(ebola, id_vars=['Date', 'Day'], var_name='status', value_name='counts')
ebola_melt = ebola_melt.dropna()
ebola_melt.pivot_table(index=['Date', 'Day'], columns='status', values='counts')
```

```python
new_ebola = pd.DataFrame({'Date':ebola.Date,
                          'Day':ebola.Day})
ebola_melt = pd.melt(ebola, id_vars=['Date', 'Day'])

variable = ebola_melt.variable
value = ebola_melt.value
status = ebola_melt.variable.str.split('_').str[0]
country = ebola_melt.variable.str.split('_').str[1]

pd.concat([new_ebola, variable, value, status.rename('status'), country.rename('country.status')], axis=1)
```
