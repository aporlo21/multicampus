```python
import numpy as np
```

```python
# 코드 사전 정의(동요)

code2idx = {'c4':0, 'd4':1, 'e4':2, 'f4':3, 'g4':4, 'a4':5, 'b4':6,
            'c8':7, 'd8':8, 'e8':9, 'f8':10, 'g8':11, 'a8':12, 'b8':13}
idx2code = {0:'c4', 1:'d4', 2:'e4', 3:'f4', 4:'g4', 5:'a4', 6:'b4',
            7:'c8', 8:'d8', 9:'e8', 10:'f8', 11:'g8', 12:'a8', 13:'b8'}

# 시퀀스 데이터 정의

seq = ['g8', 'e8', 'e4', 'f8', 'd8', 'd4', 'c8', 'd8', 'e8', 'f8', 'g8', 'g8', 'g4',
       'g8', 'e8', 'e8', 'e8', 'f8', 'd8', 'd4', 'c8', 'e8', 'g8', 'g8', 'e8', 'e8', 'e4',
       'd8', 'd8', 'd8', 'd8', 'd8', 'e8', 'f4', 'e8', 'e8', 'e8', 'e8', 'e8', 'f8', 'g4',
       'g8', 'e8', 'e4', 'f8', 'd8', 'd4', 'c8', 'e8', 'g8', 'g8', 'e8', 'e8', 'e4']
```

```python
def seq2dataset(seq, window_size):
    dataset = []
    for i in range(len(seq) - window_size): # 54 - 4 = 50 (range)
        subset = seq[i:( i + window_size + 1)]
        # i = 0 이면, ( i + window_size + 1)
        # 시작위치 window_size['g8', 'g8', 'e8', 'e8'] => ['e4']
    # print(len(seq))
        dataset.append([code2idx[item] for item in subset]) 
## 해당출력결과를 수치로 변환해서 도출하는 작업이 필요하다 
# 'g8' => 11 로 수치화 변환하는 작업부문 
    # print(dataset)
    return np.array(dataset)

```

```python
dataset = seq2dataset(seq, window_size=4) # 나비야 원곡 정보 전체
print(dataset.shape)
```

### 다층 퍼셉트론 모델링 


```python
# x, y  변수 분리, 정규화, 원핫인코딩 포함 작업

from keras.utils import np_utils
# ohe 해주는 과정 ... softmax 를 위한 전처리
max_idx_value = 13
xtrain = dataset[:, 0:4]
ytrain = dataset[:,4]
xtrain = xtrain / float(max_idx_value)
ytrain = np_utils.to_categorical(ytrain)
one_hot_vec_size = ytrain.shape[1] # (50, 12) # 12 : One hot Vector 의 크기
# 원래 7 * 2 = 14개의 음인데 사용되어진 음표는 12개이다 
# 사용되어진 음표의 종류의 갯수
```

```python
print("원핫벡터의 크기는 : ", one_hot_vec_size)
```

```python
# 다층 퍼셉트론 모델링 [Neural NEt]
from keras.models import Sequential
import keras 
from keras.models import *
from keras.layers import *
from keras.utils import *
```

```python
model=Sequential()
model.add(Dense(128, input_dim = 4, activation='relu')) # dense 3 layer
model.add(Dense(128, activation='relu')) 
model.add(Dense(one_hot_vec_size, activation='softmax')) # One hot vec size로 바껴 
```

```python
# - 처음에 음표 4개 입력 [x0,x1x,x2,x3]
# - 출력 : 128
```

```python
# 학습 알고리즘 설정 
model.compile(loss = 'categorical_crossentropy',
             optimizer = 'adam', metrics = ['accuracy'])
```

```python
class LossHistory(keras.callbacks.Callback):
    def init(self):
        self.losses=[] # self 객체 내에 변수를 가지고 있다. 
    def on_epoch_end(self, batch, logs={}): # 콜백함수 ( 알아서 탐색, 수행되는 함수 )
        self.losses.append(logs.get('loss'))
# 에폭 종료시 시스템에 의해 자동 호출되는 함수    
```

```python
history = LossHistory() # 해당 제작 함수 호출 # LossHistory 객체
# call back 함수 : 어떤 상황이 되면 시스템이 호출하는 함수 
# on ____ 함수 (~~했을때) : 전부다 어떤 상황아래 시스템이 호출하는 함수 
history.init()
```

```python
num_epochs = 1000

model.fit(xtrain, ytrain, epochs=num_epochs, batch_size=1, 
          verbose=2, callbacks=[history]) # 상황에 맞춰서 계속 수행됨
# acc : 92 
```

```python
scores = model.evaluate(xtrain, ytrain, batch_size = 1) # 리스트 구조 
print("%.2f%%"%(scores[1] * 100))
```

```python
seq_out = ['g8', 'e8', 'e4', 'f8']
pred_out = model.predict(xtrain, batch_size = 1)
# pred_out
"""
1.93987697e-15, 4.42518907e-29, 3.39105547e-24, 6.68820208e-11,
1.14002419e-09, 2.71718344e-15, 1.49340442e-15, 2.42543573e-19,
9.91866171e-01, 8.13381746e-03, 2.28937352e-14, 4.88881940e-18],
# 사용한 음 12 개 중/...
## softmax ___ 합이 1...
### 8번 index => prediction value 로 도출 
"""
np.argmax(pred_out, 1)

# 8, (첫 입력 4개에 대한 예측결과)
# 1, (그 다음 입력 4개에 대한 예측결과)
```
