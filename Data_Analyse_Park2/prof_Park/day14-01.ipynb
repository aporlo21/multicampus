```python
from sklearn.datasets import load_digits
digits=load_digits()
samples=[0,10,20,30,1,11,21,31] # 이미지번호
digits

```

```python

d=[]
for i in range(8):
    d.append(digits.images[samples[i]]) # 0,10,20,30,1,11,21,31
plt.figure(figsize=(8,2))
for i in range(8):
    plt.subplot(1,8,i+1) # 1줄 8칸 8줄
    plt.imshow(d[i], cmap=plt.cm.bone_r, interpolation='nearest') # imageshow ( 0: 4, 1: 4)
    plt.title('img {}'.format(i+1))
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])
plt.show()
#     print("-"*50)
# samples의 i 가 0일 때, 0이아닌값들을 선으로 연결시 이미지 [0] 
#
```

```python
# 2 차원이미지(8,8)를 1차원 벡터(64,1)로 펼친다
# d[0].shape
v=[]
for i in range(8):
    v.append(d[i].reshape(64,1)) # reshape // shape 바꾸기 
# np.shape(v)
plt.figure(figsize=(8,2))
for i in range(8):
    plt.subplot(1,8,i+1) # 1줄 8칸 8줄
    plt.imshow(v[i],aspect=0.2, cmap=plt.cm.bone_r, interpolation='nearest') # imageshow ( 0: 4, 1: 4)
    plt.title('img {}'.format(i+1))
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])
plt.tight_layout(w_pad=3)
plt.show()

# 1. 연습문제
# 벡터의 내적을 수행 => 가장 큰 내적을 출력(이미지 번호)
# 내적이 가장 큰다 두개의 이미지는 같은 이미지다 혹은 같은 숫자이다 . 의미 
```
