```python
# exer08-01
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
names1880=pd.read_csv("data/babynames/yob1880.txt",
                     names = ['name', 'sex', 'births'])
names1880
# 이름 성별 생년 
```

```python
years=range(1880,2011)
pieces=[]
for year in years:
    path="data/babynames/yob%d.txt" % year
    frame=pd.read_csv(path, names=['name','sex','births'])
    frame['year']=year
    pieces.append(frame)
    # year프레임을 추가해서 덧붙여줌
    
```

```python
names=pd.concat(pieces,ignore_index=True)
names
```

```python
total_births=names.pivot_table('births', index='year',
                              columns = 'sex', aggfunc = sum)
total_births
# 성별에 따른 대분류 // indexing 요소 'year'
# total_births.plot()
```

```python
def add_prop(group):
    group['prop']=group.births/group.births.sum()
    return group
# births 그룹화 births.sum 그룹화 하여 각 그룹을 나누어 준것이 비율
```

```python
names = names.groupby(['year','sex']).apply(add_prop)
print(names)
```

```python
# 연도별, 성별에 따른 빈도수가 가장 높은 이름 1000개 추출
def get_top1000(group):
    return group.sort_values(by='births', ascending=False)[:1000]
# births에 따른 분류 # 내림차순 # [:1000] 까지
```

```python
grouped = names.groupby(['year', 'sex'])
top1000 = grouped.apply(get_top1000)
top1000.index=np.arange(len(top1000))
top1000
top1000.index
```

```python
# 이름 유행분석
boys = top1000[top1000.sex=="M"]
boys
girls = top1000[top1000.sex=="F"]
girls
```

```python
total_births=top1000.pivot_table('births', index='year', 
                                 columns='name',aggfunc=sum)
total_births
```

```python
sub_set=total_births[['John','Harry','Mary','Marilyn']]
sub_set
sub_set.plot(subplots=True, figsize=(12,10), 
             title='Number of births per year')
```

```python
table=top1000.pivot_table('prop', index='year',
                         columns='sex', aggfunc=sum)
table
# prop에 대해서 top1000의 범위 내에서 100 프로 매칭되덧것이 근래에 들어서면서 
# 그 외 이름들이 많이 사용되면서 퍼센티지가 낮아지고 있는 빈도수 변화의 양상을 보임
table.plot(xticks=range(1880,2010,10), yticks=np.linspace(0,1.2,13))
# x 축 범위 // # y축 범위 
```
