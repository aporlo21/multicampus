```python
# BS - CSS selector 함수 추출
# jQuery // javascript lib 처럼 css 선택자 사용가능, 특정 요소 추출
# soup.select(<선택자>): 여러개 요소 # 지정 범위 축소 단계
# soup.select_one(<선택자>):1개 요소 
```

```python
html=\
"""
<html><body>
<h1>역삼동</h1>
<div id="myid">
<h1>멀티캠퍼스</h1>
<ul class="items">
<li>강의실</li>
<li>엘레베이터</li>
<li>식당</li>
</ui>
</div></body></html>
"""
# 꺽쇠로 묶여 있으면 태그다 ! ^^ 
soup=BeautifulSoup(html,'html.parser')
print(soup.select_one("h1"))
print(soup.select_one("div#myid h1")) # 소속이 다르다 div id 태그 내부존재 h1 '멀티캠퍼스'
print(soup.select_one("div#myid > h1"))
# (#)을 통해 구분
print(soup.select_one("li"))
print(soup.select("li")) # findall 과 같은 역할
```
